```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# separating the features and target
X = data_frame.drop(columns='label', axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# standardize the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
pip install tensorflow
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of my neural network

model = keras.Sequential([
                         keras.layers.Flatten(input_shape=(30,)), # to convert data into single dimension array
                         keras.layers.Dense(20, activation='relu'), # hidden layer with 20 neurons, we can add multiple layers
                         keras.layers.Dense(2, activation='sigmoid') # output layer with 2 neurons
                         ])
```

```python
# compiling the neural network 

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network

history= model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
# visualising accuracy and loss 

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
# visualising accuracy and loss 

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
# accuracy of the model on test data

loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
